```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import glob

%matplotlib notebook
```

```python
# Read a brain location
files = glob.glob('./*/*Location.tsv')
len(files)
```

```python
plt.figure(figsize=(10,10))
for i in range(16):
    tbl = pd.read_csv(files[i], sep='\t')
    plt.subplot(4,4,i+1)
    plt.plot(tbl.ccf_ap, label='ap')
    plt.plot(tbl.ccf_dv, label='dv')
    plt.plot(tbl.ccf_lr, label='lr')
    plt.legend()
```

```python
# Document all brain locations
area_dict = {}
for i in range(len(files)):
    session = files[i].split('\\')[1]
    tbl = pd.read_csv(files[i], sep='\t')
    area_dict[session] = tbl.allen_ontology.unique()
```
